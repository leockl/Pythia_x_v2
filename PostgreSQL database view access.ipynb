```python
import psycopg2

def fetch_and_print_rows(db_params, table_name, num_rows):
    """
    Fetches and prints a specified number of rows from a given table.
    
    Parameters:
    - db_params: A dictionary containing the database connection parameters.
    - table_name: The name of the table to fetch the rows from (as a string).
    - num_rows: The number of rows to fetch and print.
    """
    try:
        # Establishing the connection
        conn = psycopg2.connect(**db_params)
        cursor = conn.cursor()
        
        # Preparing the SQL query to fetch a limited number of rows from the table
        sql_query = f"SELECT * FROM {table_name} LIMIT {num_rows}"
        
        # Executing the SQL query
        cursor.execute(sql_query)
        
        # Fetching the specified number of records from the query
        records = cursor.fetchall()
        
        print(f"Fetching {num_rows} rows from the table '{table_name}':")
        for row in records:
            print(row)
    except Exception as e:
        print(f"An error occurred: {e}")
    finally:
        # Ensuring the cursor and connection are closed
        cursor.close()
        conn.close()

# Database connection parameters
db_params = {
    'dbname': 'postgres',
    'user': 'viewer_account',
    'password': 'Zt8O0R8i4W7cEbTXFi80SZswUXPpQann',
    'host': '136.144.62.142',
    'port': '5432'
}
```
